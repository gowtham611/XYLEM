```python
from pathlib import Path
import onnx
# Try several likely relative paths from common notebook/kernel working directories
candidates = [
    Path('assets/models/crop_prediction_model_ir9.onnx'),
    Path('..','assets','models','crop_prediction_model_ir9.onnx'),
    Path('..','..','assets','models','crop_prediction_model_ir9.onnx'),
    Path('.','assets','models','crop_prediction_model_ir9.onnx'),
    Path('lib','..','assets','models','crop_prediction_model_ir9.onnx'),
]
model_path = None
tried = []
for p in candidates:
    p_resolved = p.expanduser()
    try:
        p_resolved = p_resolved.resolve()
    except Exception:
        # resolve may fail if path components don't exist; keep the expanded path
        pass
    tried.append(str(p_resolved))
    if p_resolved.exists():
        model_path = p_resolved
        break
if model_path is None:
    raise FileNotFoundError('Model file not found. Tried: ' + ', '.join(tried))
m = onnx.load(str(model_path))
print('Using model file:', model_path)
print('Inputs:', [(i.name, [d.dim_value for d in i.type.tensor_type.shape.dim]) for i in m.graph.input])
print('Outputs:', [(o.name, [d.dim_value for d in o.type.tensor_type.shape.dim]) for o in m.graph.output])
```
